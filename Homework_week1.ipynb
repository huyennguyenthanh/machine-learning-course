## BT 4
#### Giải bài toán lớp 3 <br>
Các biến liệt kê bên dưới nằm trong đoạn từ 1 đến 9, hỏi có bn bộ giá trị thỏa mãn điều kiện sau: <br>
a + 13 * b / c + d + 12 * e - f - 11 + g * h / i - 10 = 66 <br>
Giả sử a,b,c... đôi một khác nhau.

```python
#liệt kê hoán vị
import itertools
data_4 = list(itertools.permutations([i for i in range(1,10)]))
# len(data_4) = 9!
result = 0
for i in data_4:
    if i[0] + 13 * i[1] / i[2] + i[3] + 12 * i[4] - i[5] - 11 + i[6] * i[7] / i[8] - 10 == 66:
        result +=1
print('The number of sets of values satisfying the experssion is',result)    
```

## BT 5
Nếu lần lượt cho các giá trị trong bảng chữ cái tiếng Anh như sau: a = 1, b = 2, c = 3, ....
Hỏi các từ sau có giá trị là bao nhiêu: python, patience, documents, students, 
homework, practice, success, english, university, congratulation

```python
alphabet = list('abcdefghijklmnopqrstuvwxyz')
digit = [i for i in range(1,27)]  
dict_5 = dict(zip(alphabet, digit))
print(dict_5)
```

```python
data_5 = ['abc', 'python', 'patience', 'documents', 'students', 'homework', 'practice', 
          'success', 'english', 'university', 'congratulation']
for word in data_5:
    value = 0
    print(word , end = " = ")
    for i in word:
        value += dict_5[i]
        print(dict_5[i] , end = (" + " if i != word[len(word) - 1] else " "))
    print("=",value)       
```

## BT6
Viết code sinh các số fibonanci dưới 1000, lưu giữ kq vào 1 list.

```python
result = [1,1]
i = 0
while True:
    c = result[i] + result[i+1]
    if c > 1000:
        break
    result.append(c)
    i+=1
print(result)
```

## BT7
Viết code sinh các số nguyên tố dưới 10000, lưu giữ kq vào 1 list.

```python
import numpy as np
def is_prime(n):
    for i in range(2,int(np.sqrt(n)+1)):
        if n%i == 0:
            return False
    return True
```

```python
result = [2]+ [i for i in range(3,10000,2) if is_prime(i) ]
result
```
